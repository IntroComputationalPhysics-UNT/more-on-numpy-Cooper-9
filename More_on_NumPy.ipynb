```python
# COMPLETE THIS TASK
# create a vectorized function that calculates the total energy (T+U; kinetic + potential energy) of a harmonic oscillator
```

```python
def total_harmonic_oscillator_energy(x, v, m, k):
    """
    Calculates the total energy (kinetic + potential) of a harmonic oscillator.

    Parameters:
    x (array_like): Displacement from equilibrium.
    v (array_like): Velocity.
    m (float): Mass of the oscillating object.
    k (float): Spring constant.

    Returns:
    array_like: Total energy of the harmonic oscillator.
    """
    kinetic_energy = 0.5 * m * v**2
    potential_energy = 0.5 * k * x**2
    total_energy = kinetic_energy + potential_energy
    return total_energy

# Define parameters
mass = 1.0  # kg
spring_constant = 10.0  # N/m

# Create arrays for displacement and velocity
displacements = np.array([0.1, 0.2, 0.3, -0.1, -0.2]) # meters
velocities = np.array([0.5, 0.3, 0.1, -0.4, -0.6]) # m/s

# Calculate total energy
energies = total_harmonic_oscillator_energy(displacements, velocities, mass, spring_constant)

print("Displacements (x):", displacements)
print("Velocities (v):", velocities)
print("Total Energies:", energies)
```

```python
# COMPLETE THESE TASKS
# Show that U@U.T and U.T@U are 2x2 identity matrices
# Note that normally we would need to use the `.conj().T` attribute to find the conjugate transpose matric, but our matrix is real.

U = np.array([[1,0],[0,1]])
U_T = U.T

print(U @ U.T)
print(U.T @ U)
```

```python
# COMPLETE THIS TASK
# From the eigenvalue equation M v = λ v, it follows that for U = [v1, v2], then M U = λ U.
# Because λ is diagonal and U.T@U = np.eye(2), it follows that λ = U.T @ K @ U.
# Confirm this behavior.


# U is the matrix whose columns are the eigenvectors
U_eigenvectors = vecs

# Form the diagonal matrix of eigenvalues (lambda_matrix)
lambda_matrix = np.diag(vals)

# Calculate U.T @ K @ U
computed_lambda = U_eigenvectors.T @ K @ U_eigenvectors

print("Original Eigenvalues (diagonal matrix):\n", lambda_matrix)
print("\nComputed U.T @ K @ U:\n", computed_lambda)

# Confirm that the computed_lambda is approximately equal to lambda_matrix
# np.allclose is used for floating-point comparisons
is_confirmed = np.allclose(computed_lambda, lambda_matrix)
print(f"\nDoes U.T @ K @ U approximately equal the diagonal eigenvalue matrix? {is_confirmed}")

# Also confirm that U.T @ U is approximately an identity matrix (for orthogonal eigenvectors)
# This is true for symmetric matrices like K
identity_check = U_eigenvectors.T @ U_eigenvectors
print("\nU.T @ U (Identity check):\n", identity_check)
print(f"Is U.T @ U approximately an identity matrix? {np.allclose(identity_check, np.eye(K.shape[0]))}")
```

```python
k = 2450
K = np.array([[k, -k], [-k, k]])

vals, vecs = np.linalg.eigh(K)

omega = np.sqrt(vals)

print("Force constant (k):", k)
print("K matrix:\n", K)
print("Eigenvalues (vals):", vals)
print("Eigenvectors (vecs):\n", vecs)
print("Vibrational frequencies (omega):", omega)
```

### Explanation of Potential Energy Formula

The potential energy formula for a system with displacements `x` and force constant matrix `K` is given by `V = 0.5 * x.T @ K @ x`. In the context of a diatomic molecule, `x = (x1, x2)` represents the displacements of the two atoms from their equilibrium positions. `x.T` is the transpose of the displacement vector.

This formula calculates the total potential energy stored in the system due to the elastic deformation (stretching or compression) of the bond between the two atoms. The `K` matrix captures the interatomic forces. For a diatomic molecule, the `K` matrix describes how the potential energy changes with the relative displacement of the two atoms. A higher potential energy indicates a greater 'strain' or displacement from the equilibrium configuration.

### Interpretation of the First Eigenvalue and Eigenvector

**Eigenvalue 1 (λ₁):** `vals[0] = 0` (or `9.99658224e-17` from the previous general problem, but for this specific diatomic molecule it's exactly 0).

**Corresponding Eigenvector 1 (v₁):** `vecs[:, 0] = [-0.70710678, -0.70710678]`

**Vibrational Frequency 1 (ω₁):** `omega[0] = np.sqrt(0) = 0`

**Physical Interpretation:**

1.  **Motion of Atoms:** The eigenvector `[-0.70710678, -0.70710678]` indicates that both atoms are displaced in the same direction and by the same magnitude. For instance, if the first atom moves left by 0.707 units, the second atom also moves left by 0.707 units (or vice versa, depending on the chosen direction for the eigenvector, as `[0.707, 0.707]` would also be a valid eigenvector). This represents a **pure translational motion** of the entire molecule as a single unit, where the bond length between the two atoms remains unchanged.

2.  **Vibrational Frequency and Potential Energy:** The corresponding vibrational frequency `ω₁ = 0` is crucial. A zero frequency implies that this mode requires **no energy** to be excited and does not involve any oscillation. If the molecule is translated, there is no change in the potential energy. This is consistent with the potential energy formula `V = 0.5 * x.T @ K @ x`. If `x` is proportional to `v₁`, then `K v₁ = λ₁ v₁ = 0 * v₁ = 0`, so `V = 0.5 * x.T @ 0 = 0`. This means that translating the molecule as a whole does not stretch or compress the bond, and thus stores no potential energy in the internal spring.

3.  **Mode Type:** This mode is a **translational mode**, not a vibrational mode. It describes the motion of the center of mass of the diatomic molecule without any internal deformation.

### Interpretation of the Second Eigenvalue and Eigenvector

**Eigenvalue 2 (λ₂):** `vals[1] = 4900`

**Corresponding Eigenvector 2 (v₂):** `vecs[:, 1] = [-0.70710678, 0.70710678]`

**Vibrational Frequency 2 (ω₂):** `omega[1] = np.sqrt(4900) = 70`

**Physical Interpretation:**

1.  **Motion of Atoms:** The eigenvector `[-0.70710678, 0.70710678]` indicates that the two atoms are displaced in **opposite directions** but with the **same magnitude**. If the first atom moves left by 0.707 units, the second atom moves right by 0.707 units. This represents a **relative motion** where the distance between the two atoms changes (the bond is stretched or compressed).

2.  **Vibrational Frequency and Potential Energy:** The corresponding vibrational frequency `ω₂ = 70` is non-zero. A non-zero frequency signifies that this mode is a true oscillation and requires energy to be excited. When the atoms move according to this eigenvector, the bond length changes, leading to a change in the potential energy of the system. This is consistent with `V = 0.5 * x.T @ K @ x`. If `x` is proportional to `v₂`, then `K v₂ = λ₂ v₂`. Since `λ₂` is non-zero, the potential energy `V` will also be non-zero, indicating energy stored in the deformation.

3.  **Mode Type:** This mode is a **vibrational mode**. It describes the internal oscillation of the diatomic molecule, where the atoms move relative to each other, stretching and compressing the bond.

### Summary of Normal Modes for the Diatomic Molecule

For the diatomic molecule described by the `K` matrix `[[k, -k], [-k, k]]` (with `k = 2450`), we have identified two normal modes:

1.  **First Normal Mode (Translational Mode):**
    *   **Eigenvalue (λ₁):** `0`
    *   **Eigenvector (v₁):** `[-0.70710678, -0.70710678]` (or `[1, 1]` normalized)
    *   **Vibrational Frequency (ω₁):** `0`
    *   **Physical Meaning:** This mode represents the **center-of-mass translation** of the entire diatomic molecule. Both atoms move together in the same direction and with the same magnitude. Since there is no change in the relative distance between the atoms, the bond length remains constant, and therefore, no potential energy is stored in this motion (`V = 0`). The zero vibrational frequency signifies that this mode is not an oscillation but a free translation, which does not require energy to be excited.

2.  **Second Normal Mode (Vibrational Mode):**
    *   **Eigenvalue (λ₂):** `4900`
    *   **Eigenvector (v₂):** `[-0.70710678, 0.70710678]` (or `[-1, 1]` normalized)
    *   **Vibrational Frequency (ω₂):** `70`
    *   **Physical Meaning:** This mode represents the **internal vibration** of the diatomic molecule. The two atoms move in opposite directions, stretching and compressing the bond. This relative motion changes the distance between the atoms, leading to a storage of potential energy (`V > 0`). The non-zero vibrational frequency (70 rad/s) indicates that this is a true oscillation, and energy is required to excite this mode. This is the only true vibrational mode for a linear diatomic molecule.
